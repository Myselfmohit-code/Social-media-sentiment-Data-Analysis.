```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment Label', data=df, palette='coolwarm')
plt.title('Distribution of Sentiment Labels')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sentiment Label', y='Number of Likes', data=df, palette='coolwarm')
plt.title('Number of Likes by Sentiment Label')
plt.show()
```

```python
sns.pairplot(df[['Number of Likes', 'Number of Shares', 'Number of Comments']], diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='User Follower Count', y='Number of Likes', hue='Post Type', data=df, palette='coolwarm', alpha=0.6)
plt.title('Number of Likes vs Follower Count by Post Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Sentiment Label', hue='Post Type', data=df, palette='coolwarm')
plt.title('Distribution of Post Types by Sentiment Label')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Sentiment Label', y='Number of Shares', data=df, palette='Set2')
plt.title('Distribution of Number of Shares by Sentiment Label')
plt.show()
```
